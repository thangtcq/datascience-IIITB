**Model Interpretation**

The output summary table  will provide the features used for building model along with coefficient of each of the feature and their p-value. The p-value in a logistic regression model is used to assess the statistical significance of each coefficient. Lesser the p-value, more significant the feature is in the model.

A positive coefficient will indicate that an increase in the value of feature would increase the odds of the event occurring. On the other hand, a negative coefficient means the opposite, i.e,  an increase in the value of feature would decrease the odds of the event occurring.

7.3.1 Plot ROC curve <font color="red">[3 Marks]</font>

```python
# Define ROC function
from sklearn.metrics import roc_curve, auc

def plot_roc_curve(y_true, y_scores):
    fpr, tpr, thresholds = roc_curve(y_true, y_scores)

    roc_auc = auc(fpr, tpr)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange',
             lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--',
             label='Random Guess')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.show()
    
    return fpr, tpr, thresholds, roc_auc

fpr, tpr, thresholds, roc_auc = plot_roc_curve(results_df['Actual_Numeric'], results_df['Predicted_Probability'])

print("AUC:", roc_auc)
```

7.3.2 Predict on training set at various probability cutoffs <font color="red">[1 Mark]</font>

```python
# Predict on training data by creating columns with different probability cutoffs to explore the impact of cutoff on model performance
cutoff_values = [0.4, 0.45, 0.5, 0.55, 0.6]

for cutoff in cutoff_values:
    column_name = f"Predicted_{cutoff}"
    results_df[column_name] = results_df["Predicted_Probability"].apply(lambda x: 1 if x > cutoff else 0)

print("Training data with predictions for various cutoffs:")
print(results_df.head())
```

7.3.3 Plot for accuracy, sensitivity, specificity at different probability cutoffs <font color="red">[2 Marks]</font>

```python
# Create a DataFrame to see the values of accuracy, sensitivity, and specificity at different values of probability cutoffs
from sklearn.metrics import accuracy_score, confusion_matrix

cutoff_values = [0.4, 0.45, 0.5, 0.55, 0.6]

metrics_results = []

for cutoff in cutoff_values:
    pred = results_df["Predicted_Probability"].apply(lambda x: 1 if x >= cutoff else 0)
    
    accuracy = accuracy_score(results_df["Actual_Numeric"], pred)
    
    tn, fp, fn, tp = confusion_matrix(results_df["Actual_Numeric"], pred).ravel()
    
    sensitivity = tp / (tp + fn) if (tp + fn) != 0 else 0
    
    specificity = tn / (tn + fp) if (tn + fp) != 0 else 0
    
    metrics_results.append({
        "Cutoff": cutoff,
        "Accuracy": accuracy,
        "Sensitivity": sensitivity,
        "Specificity": specificity
    })

metrics_df = pd.DataFrame(metrics_results)

print("Performance metrics at different probability cutoffs:")
print(metrics_df)
```

```python
# Plot accuracy, sensitivity, and specificity at different values of probability cutoffs
plt.figure(figsize=(10, 6))

plt.plot(metrics_df['Cutoff'], metrics_df['Accuracy'], marker='o', label='Accuracy')
plt.plot(metrics_df['Cutoff'], metrics_df['Sensitivity'], marker='s', label='Sensitivity')
plt.plot(metrics_df['Cutoff'], metrics_df['Specificity'], marker='^', label='Specificity')

plt.title('Performance Metrics at Different Probability Cutoffs')
plt.xlabel('Probability Cutoff')
plt.ylabel('Metric Value')

plt.legend()
plt.grid(True)
plt.show()
```

7.3.4 Create a column for final prediction based on the optimal cutoff <font color="red">[2 Marks]</font>

```python
# Create a column for final prediction based on the optimal cutoff
metrics_df['Youden_Index'] = metrics_df['Sensitivity'] + metrics_df['Specificity'] - 1

optimal_cutoff = metrics_df.loc[metrics_df['Youden_Index'].idxmax(), 'Cutoff']
print("Optimal Cutoff Value:", optimal_cutoff)

results_df['Final_Prediction'] = results_df['Predicted_Probability'].apply(lambda x: 1 if x >= optimal_cutoff else 0)

print("DataFrame with final predictions based on the optimal cutoff:")
print(results_df[['Predicted_Probability', 'Final_Prediction']].head())
```
